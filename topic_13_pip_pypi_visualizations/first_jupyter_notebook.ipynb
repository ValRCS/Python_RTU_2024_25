## Introduction

Jupyter Notebooks are a great tool for data analysis and visualization. They allow you to write and execute code, display the results, and add explanations in markdown all in one document. This makes it easy to share your work with others and to reproduce your results.

## Markdown (Text) Cells

Markdown cells are used to write explanations and notes. You can format the text using markdown syntax. For example, you can create headers using `#`, `##`, `###`, etc., make text bold or italic using `**` and `*`, create lists using `*` or `-`, and insert links using `[text](url)`.

```python
# variables persist as long as kernel is running and not restarted
# variables are shared between cells
print(f'My variable is still here: {my_variable}')	# My variable is still here: 42
my_variable += 10
print(f'my_variable: {my_variable}')	# my_variable: 52
```

## Using Matplotlib in Jupyter Notebooks

Let's look at a simple example of using Matplotlib in a Jupyter Notebook. First, we need to import the `matplotlib` library and specify that we want to display the plots in the notebook.

Documentation: [Matplotlib](https://matplotlib.org/)


```python
# first we need to install the package if we don't have it
try:
    import matplotlib.pyplot as plt
    print("Great success: matplotlib is installed")
except ImportError:
    print("matplotlib is not installed")
    print("Install it with: pip install matplotlib")
```

### Plotting a simple square and cube function



```python
# let's get x from 1 to 20
x = range(1, 21)
y = [value ** 2 for value in x]
y2 = [value ** 3 for value in x]

plt.plot(x, y, label='x^2', color='red')
plt.plot(x, y2, label='x^3', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x^2 and x^3')
plt.legend() # show explanation of the lines
plt.show() # this gives you static plot below
```

```python
# let's make the same plot but draw grid and draw green arrow to 10,1000

plt.plot(x, y, label='x^2', color='red')
plt.plot(x, y2, label='x^3', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x^2 and x^3')
plt.legend() # show explanation of the lines
plt.grid() # draw grid
plt.annotate('10, 1000', xy=(10, 1000), xytext=(10, 2000), arrowprops=dict(facecolor='green', shrink=0.05))
plt.show() # this gives you static plot below
```
